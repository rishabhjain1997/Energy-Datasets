```python
generation_df = pd.read_csv('Final Datasets/energy_generation.csv')
generation_df = generation_df[(generation_df['timestamp'] <= '2018-01-01') & (generation_df['timestamp'] >= '2017-01-01')].set_index('timestamp')
```

```python
generation_df['Energy Generation'] = generation_df['Wind Energy'] + generation_df['Solar Energy']
#generation_df = generation_df.set_index('timestamp')
generation_df
```

```python
generation_df['Hours'] = pd.Series([i for i in range(8760)]).values
generation_df.set_index('Hours')
```

```python
consumption_df = pd.read_csv('Final Datasets/energy_consumption.csv')
consumption_df = consumption_df[(consumption_df['timestamp'] <= '2018-01-01 00:00:00') & (consumption_df['timestamp'] >= '2017-01-01')].set_index('timestamp')
consumption_df.drop('Unnamed: 0', axis = 1, inplace = True)
#consumption_df['timestamp'] =consumption_df.index
consumption_df = consumption_df.groupby(['timestamp']).sum()
consumption_df['Hours'] = pd.Series([i for i in range(8760)]).values
consumption_df = consumption_df[['Hours', 'energy_kWh']].set_index('Hours')
consumption_df
```

```python
generation_df['Energy Consumption'] = consumption_df['energy_kWh'].values
```

```python
generation_df['Deficit'] =  -(generation_df['Energy Generation']  - generation_df['Energy Consumption'] )
```

```python
generation_df.to_csv('Microgrid_Dataset.csv')
```
