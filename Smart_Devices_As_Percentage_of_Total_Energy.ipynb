```python
import numpy as np
from statistics import mean
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import pandas as pd
import seaborn as sns
from tslearn.clustering import TimeSeriesKMeans
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, SimpleRNN 
from tensorflow.keras import regularizers
from tensorflow.keras.layers import Dense, Dropout
import tensorflow as tf
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from tensorflow import keras
import seaborn as sns
import pickle
import functools
import fastdtw
import scipy.spatial.distance as sd
#from datetime import datetime
consumption_df = pd.read_csv('Final Datasets/energy_consumption.csv')
clustered_df = pd.read_csv('Tables/clustered_AfterStandardization_K=5.csv')
#consumption_df
```

```python
consumption_df = consumption_df.drop(['Unnamed: 0', 'timestamp'], axis=1)
```

```python
house_df = consumption_df.groupby(['House Number']).sum()
for smart_feature in ['electronics workbench energy',
       'network equipment energy', 'home office energy',
       'RAE home office energy', 'RAE Mobile Phone Charger Energy']:
    house_df[smart_feature] = (house_df[smart_feature]/house_df['energy_kWh']) * 100
house_df = house_df.drop(['energy_kWh'], axis=1)
print('Smart devices as percentage of total energy consumption:')
house_df
```

```python
clustered_df.rename(columns = {'House':'House Number'}, inplace = True)
consumption_df = pd.merge(clustered_df, consumption_df, on='House Number')
```

```python
cluster_df = consumption_df.drop(['House Number'], axis=1)
#cluster_df
```

```python
cluster_df = cluster_df.groupby(['Cluster']).sum()
for smart_feature in ['electronics workbench energy',
       'network equipment energy', 'home office energy',
       'RAE home office energy', 'RAE Mobile Phone Charger Energy']:
    cluster_df[smart_feature] = (cluster_df[smart_feature]/cluster_df['energy_kWh']) * 100
cluster_df = cluster_df.drop(['energy_kWh'], axis=1)
print('Smart devices as percentage of total energy consumption:')
cluster_df
```
